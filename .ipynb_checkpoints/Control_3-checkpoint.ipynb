```python
# 1) Integrar los archivos que est√°n el folder comprimido
# primero subo los archivos
import pandas as pd
from google.colab import files
uploaded = files.upload()
```

```python
# Asigno un nombre a cada archivo, pero para el level1 solo
# quiero el encabezado a partir de la fila 4
level1 = pd.read_excel('level1.xlsx', header = [4])
level2 = pd.read_excel('level2.xlsx')
level3 = pd.read_excel('level3.xlsx')
level4 = pd.read_excel('level4.xlsx')

# Cada documento tiene columnas en com√∫n, solo en 3 de ellas (level1,level2,level3) tendr√≠amos que
# quitar dos filas que no correspondr√≠an al encabezado
# la fila de 2021 ser√° agregada despu√©s
# level 4 ser√≠a la especie de "base" para este caso
level4
```

```python
# Para level1:
level1 = level1.drop(level1.index[[0,1]])
level1.reset_index(drop=True, inplace=True)
level1
```

```python
# Para el level2:
level2 = level2.drop(level2.index[[0,1]])
level2.reset_index(drop=True, inplace=True)
level2
```

```python
# Para el level3:
level3 = level3.drop(level3.index[[0,1]])
level3.reset_index(drop=True, inplace=True)
level3
```

```python
# antes de integrar las tablas, debemos de encontrar las columnas en com√∫n y colocarlas como una sola para todas
level1_columns = set(level1.columns)
level2_columns = set(level2.columns)
level3_columns = set(level3.columns)
level4_columns = set(level4.columns)
common_columns = list(level1_columns & level2_columns & level3_columns & level4_columns)
common_columns
```

```python
# estas columnas en com√∫n se agregar a las tablas y las integro
level1 = level1[common_columns]
level2 = level2[common_columns]
level3 = level3[common_columns]
level4 = level4[common_columns]
data_integrada = pd.concat([level1,level2,level3,level4],axis=0)
data_integrada
```

```python
# cambiamos los nombres de los encabezados de tal manera que no hayan espacios, y tambi√©n reseteamos el orden de las columnas
data_integrada = data_integrada.rename(columns={'Gross national income (GNI) per capita':'GNI_per_capita_2021',
                                                'Human Development Index (HDI) ':'HDI_2021',
                                                'Life expectancy at birth':'LEAB_2021',
                                                'HDI rank':'HDI_rank_2021','Country':'Country_2021',
                                                'GNI per capita rank minus HDI rank':'GNI_per_capita_rank_minus_HDI_rank_2021',
                                                'Expected years of schooling':'EYOS_2021',
                                                'Mean years of schooling':'MYOS_2021'})
cambiar_orden = ['HDI_rank_2021','Country_2021','HDI_2021','LEAB_2021','EYOS_2021','MYOS_2021',
                 'GNI_per_capita_2021','GNI_per_capita_rank_minus_HDI_rank_2021']
data_integrada = data_integrada[cambiar_orden]
data_integrada
```

```python
from thefuzz import process
def fuzzy_match(x, choices, scorer, cutoff):
  return process.extractOne(x, choices=choices, scorer=scorer, score_cutoff=cutoff)
data_integrada['matched_country'] = data_integrada['Country_2021'].apply(fuzzy_match,
                                                                         args=(country_codes['Country'],process.fuzz.token_sort_ratio, 85))
data_integrada['matched_country'] = data_integrada['matched_country'].apply(lambda x: x[0] if x else None)
merged_df = pd.merge(data_integrada, country_codes, left_on='matched_country', right_on='Country', how='left')
merged_df
```

```python
import pandas as pd
from google.colab import files
uploaded3 = files.upload()
```

```python
fallecidos_covid = pd.read_csv('fallecidosCOVID.csv',sep=';')
fallecidos_covid
```

```python
# vemos que las fechas est√°n en formato YYMMDD pero los queremos en formato %Y-%M-%D
fallecidos_covid['FECHA_FALLECIMIENTO'] = pd.to_datetime(fallecidos_covid['FECHA_FALLECIMIENTO'], format='%Y%m%d')
# luego queremos poder extraer el a√±o, mes y d√≠a de cada fecha
fallecidos_covid['A√±o'] = fallecidos_covid['FECHA_FALLECIMIENTO'].dt.year
fallecidos_covid['Mes'] = fallecidos_covid['FECHA_FALLECIMIENTO'].dt.month
fallecidos_covid['D√≠a'] = fallecidos_covid['FECHA_FALLECIMIENTO'].dt.day
fallecidos_covid['Mes_nombre'] = fallecidos_covid['Mes'].apply(lambda x: {
    1: 'Enero',
    2: 'Febrero',
    3: 'Marzo',
    4: 'Abril',
    5: 'Mayo',
    6: 'Junio',
    7: 'Julio',
    8: 'Agosto',
    9: 'Septiembre',
    10: 'Octubre',
    11: 'Noviembre',
    12: 'Diciembre'
}[x])
fallecidos_covid
```

```python
# ahora genero mi tabla con el reporte mensual de fallecidos
reporte_mensual_fallecidos = fallecidos_covid.groupby(['A√±o','Mes','Mes_nombre']).size().reset_index(name='Fallecidos')
nuevo_orden_reporte = ['A√±o','Mes','Mes_nombre','Fallecidos']
reporte_mensual_fallecidos = reporte_mensual_fallecidos[nuevo_orden_reporte]
reporte_mensual_fallecidos
```

```python
# 4)  Use el mismo archivo anterior. Esta vez haga un reporte de fallecidos por mes, pero el tipo de
#     muerte debe estar en columnas separadas.
# ahora quiero sacar un reporte de fallecidos por mes pero incluyendo el tipo de muerte
reporte_fallecidos_causa = fallecidos_covid.groupby(['A√±o', 'Mes', 'Mes_nombre', 'CLASIFICACION_DEF'])['CLASIFICACION_DEF'].count().reset_index(name='Fallecidos')
reporte_fallecidos_causa = reporte_fallecidos_causa.pivot_table(index=['A√±o', 'Mes', 'Mes_nombre'],
                                                               columns='CLASIFICACION_DEF',
                                                               values='Fallecidos',
                                                               fill_value=0)
reporte_fallecidos_causa
```

```python
# editamos la tabla
reporte_fallecidos_causa = reporte_fallecidos_causa.rename(columns={'Criterio SINADEF':'C_Sinadef',
                                                'Criterio cl√≠nico':'C_Cl√≠nico',
                                                'Criterio investigaci√≥n Epidemiol√≥gica':'C_I_Epidemiol√≥gica',
                                                'Criterio nexo epidemiol√≥gico':'C_N_Epidemiol√≥gico','Criterio radiol√≥gico':'C_Radiol√≥gico',
                                                'Criterio serol√≥gico':'C_Serol√≥gico',
                                                'Criterio virol√≥gico':'C_Virol√≥gico',
                                                'NotiCovid':'NotiCovid','sinadef':'Sinadef'})
reporte_fallecidos_causa
```

```python
columnas_numericas = ['C_Sinadef','C_Cl√≠nico','C_I_Epidemiol√≥gica','C_N_Epidemiol√≥gico','C_Radiol√≥gico',
                      'C_Serol√≥gico','C_Virol√≥gico','NotiCovid','Sinadef']
reporte_fallecidos_causa[columnas_numericas] = reporte_fallecidos_causa[columnas_numericas].astype(int)
reporte_fallecidos_causa
```
